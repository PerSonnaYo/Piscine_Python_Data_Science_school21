```python
import pandas as pd
```

```python
df = pd.read_csv('..\\data\\auto.csv', index_col='ID')
```

```python
df.count()
```

## дропаем дубликаты и считаем количество

```python
df = df.drop_duplicates(subset=['CarNumber'	,'Make_n_model', 'Fines'], keep='last')
```

```python
df.count()
```

## Работаем с заполенениями

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=1, thresh=500)
```

```python
df.isnull().sum()
```

```python
df['Refund'] = df['Refund'].fillna(method='pad')
df.isnull().sum()
df2 = df[df['Fines'].isna()]
df2
```

```python
df1 = df[df['Fines'].notnull()]
mean_fine = df1['Fines'].mean()
df['Fines'] = df['Fines'].fillna(mean_fine)
df.isnull().sum()
```

## Создаем новые колонки Make and module

```python
df['Make'] = df.apply(lambda row: row['Make_n_model'].split()[0], axis=1)
df['Model'] = df.apply(lambda row: row['Make_n_model'].split()[1]
                       if len(row['Make_n_model'].split()) > 1 else df2['Fines'].iloc[0],
                       axis=1)
df = df.drop(labels=['Make_n_model'], axis=1)
# df = df[df['Make'] != df['Model']]
df.count()
```

```python
df['Fines'].mean()
```

## Сохраняем в json

```python
df.to_json('..\\data\\auto.json', orient='records')
```
