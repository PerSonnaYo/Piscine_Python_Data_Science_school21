## Группировки с номерами машин

```python
df1 = df.groupby(['CarNumber']).agg({'Fines' : ['count']})
df1.columns = df1.columns.droplevel(1)
df1.rename(columns={df1.columns[0]:"Number_of_Fines"}, inplace=True)
df1.sort_values(by='Number_of_Fines', ascending=False, inplace=True)
df1.head()
```

## Есть ли другие марки у одного номера

```python
df1 = df.groupby(['CarNumber', 'Model']).count()
df1.reset_index(level='Model', inplace = True)
df1 = df1.groupby(['CarNumber']).agg({'Model' : ['count']})
df1.columns = df1.columns.droplevel(1)
df1.rename(columns={df1.columns[0]:"Count_of_Models"}, inplace=True)
df1.sort_values(by='Count_of_Models', ascending=False, inplace=True)
df1
```
